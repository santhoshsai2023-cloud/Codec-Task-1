```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import os

# The file path has been provided by the user
file_name = "/content/Gold Price for Sales Forecasting.csv"

data = pd.read_csv(file_name)
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Inspect column names to find the correct sales column
# print(data.columns) # This would show the actual column names

# Assuming 'Gold Price ()' is the correct column based on inspection of `data` variable in kernel state
# Clean and convert 'Gold Price ()' to numeric
data['Gold Price ()'] = data['Gold Price ()'].str.replace('', '', regex=False).str.replace(',', '', regex=False).astype(float)

sales = data['Gold Price ()']

# Sort the sales Series by its DatetimeIndex to ensure monotonic order
sales = sales.sort_index()

plt.figure(figsize=(10,5))
plt.plot(sales)
plt.title("Original Sales Trend")
plt.show()

# Adjust train/test split dates to be within the actual data range (data starts in 2026)
train = sales[:'2026-01-30']
test = sales['2026-01-31':]

model = ARIMA(train, order=(1,1,1))
model_fit = model.fit()

forecast = model_fit.forecast(steps=len(test))

plt.figure(figsize=(10,5))
plt.plot(train, label='Train')
plt.plot(test, label='Actual')
plt.plot(test.index, forecast, label='Forecast')
plt.legend()
plt.title("Forecast vs Actual Sales")
plt.show()

mse = mean_squared_error(test, forecast)
rmse = np.sqrt(mse)

print("RMSE:", rmse)

future_forecast = model_fit.forecast(steps=25)

plt.figure(figsize=(10,5))
plt.plot(sales, label='Historical')
plt.plot(future_forecast, label='Future Forecast')
plt.legend()
plt.show()
```
